```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url_conf = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
url_dead = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv"
url_recov = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv"
df_conf = pd.read_csv(url_conf)
df_dead = pd.read_csv(url_dead)
df_recov = pd.read_csv(url_recov)
```

```python
thai_df = df_conf[df_conf["Country/Region"]=="Thailand"]
thai_df
```

```python
covid_df = pd.DataFrame()
covid_df['datetime'] = thai_df.columns[4:]
covid_df['confirmed'] = df_conf[df_conf["Country/Region"]=="Thailand"].loc[0][4:].values
covid_df['deaths']    = df_dead[df_dead["Country/Region"]=="Thailand"].loc[0][4:].values
covid_df['recovered'] = df_recov[df_recov["Country/Region"]=="Thailand"].loc[0][4:].values
```

```python
covid_df.tail()
```

```python
ydata = covid_df['confirmed'].values.astype(int)
xdata = covid_df.index.values.astype(int)
plt.plot(xdata, ydata, 'o', color ='blue', label ="data")
plt.show()
```

```python
import numpy.polynomial.polynomial as poly

x = list(xdata)
y = list(ydata)
len_pred = 5

x_pred = []
x_pred.extend(x)
x_pred.extend(list(range(max(x)+1,max(x)+1+len_pred)))
coefs = poly.polyfit(x, y, 6)
y_pred = poly.polyval(x_pred, coefs)
df_pred = pd.DataFrame()
df_pred['datetime'] = x_pred
df_pred['confirmed_predict'] = y_pred.astype(int)
```

```python
plt.plot(x, y, 'o', color ='blue', label="Original covid19 confirmed")
plt.plot(x_pred, y_pred, color ='red', label = "Predict covid19 confirmed")
plt.legend(loc='upper left')
plt.show()
```
